```python
# Define Panel visualization functions
def neighbourhood_map():
    """Neighbourhood Map"""
   
    # YOUR CODE HERE!
    
    area_plot = px.scatter_mapbox(
            df_mapbox,
            lat="lat",
            lon="lon",
            size="average_house_value",
            color="average_house_value",
            zoom=8)
    return area_plot


def create_bar_chart(data, title, xlabel, ylabel, color):
    """
    Create a barplot based in the data argument.
    """
    barplot = data.hvplot.bar(title=title, xlabel = xlabel, ylabel=ylabel, color=color, rot=90,
                             height=500,width=1000, yformatter=BasicTickFormatter(use_scientific=False))
    
    return barplot
    
    # Create a bar chart per year to show the number of dwelling types

    
# Bar chart for 2001
# YOUR CODE HERE!
plt_2001 = create_bar_chart(data_dwelling_types.loc[2001], title='Dwelling Types in Toronto in 2001', xlabel = '2001'
                 , ylabel='Dwelling Type Units', color='red')

# Bar chart for 2006
# YOUR CODE HERE!
plt_2006 = create_bar_chart(data_dwelling_types.loc[2006], title='Dwelling Types in Toronto in 2006', xlabel = '2006'
                 , ylabel='Dwelling Type Units', color='blue')
# Bar chart for 2011
# YOUR CODE HERE!
plt_2011 = create_bar_chart(data_dwelling_types.loc[2011], title='Dwelling Types in Toronto in 2011', xlabel = '2011'
                 , ylabel='Dwelling Type Units', color='yellow')

# Bar chart for 2016
# YOUR CODE HERE!
plt_2016 = create_bar_chart(data_dwelling_types.loc[2016], title='Dwelling Types in Toronto in 2016', xlabel = '2016'
                 , ylabel='Dwelling Type Units', color='green')

plt_plots = pn.Column('2',plt_2001,plt_2006,plt_2011,plt_2016)
    
    
    
    # YOUR CODE HERE!

def create_line_chart(data, title, xlabel, ylabel, color):
    """
    Create a line chart based in the data argument.
    """
    myplot = data.hvplot.line(title=title, xlabel=xlabel, ylabel=ylabel, color=color)
    return myplot
    
    
plot_owned = create_line_chart(data_shelter['shelter_costs_owned'], 'Avg Monthly Shelter Cost for Owned Dwellings in Toronto'
                                            , 'Year', 'Avg Monthly Shelter Cost', 'purple')
    
plot_rented = create_line_chart(data_shelter['shelter_costs_rented'], 'Avg Monthly Shelter Cost for Rented Dwellings in Toronto'
                                            , 'Year', 'Avg Monthly Shelter Cost', 'yellow')
owned_rent_plot = plot_owned + plot_rented
   
    
    # YOUR CODE HERE!

def average_house_value():
    """Average house values per year."""
    
    # YOUR CODE HERE!

def average_value_by_neighbourhood():
    """Average house values by neighbourhood."""
    
    # YOUR CODE HERE!
    price_by_neig_plot = to_data.hvplot(kind='line', x='year', y='average_house_value', ylabel='Avg. House Value', groupby='neighbourhood'
              ,height=500,width=500, yformatter=BasicTickFormatter(use_scientific=False))

    return price_by_neig_plot

    
def number_dwelling_types():
    """Number of dwelling types per year"""
    
    # YOUR CODE HERE!
    dwelling_plot = to_data.hvplot(kind='bar', x='year', y=dwelling_types, xlabel='Year', ylabel='Dwelling Type Units', groupby='neighbourhood'
              , rot=90, height=500,width=800, yformatter=BasicTickFormatter(use_scientific=False))
    return dwelling_plot
    
    
def average_house_value_snapshot():
    """Average house value for all Toronto's neighbourhoods per year."""
    to_data_resetindex = to_data.reset_index()
    average_house_value_snapshot_plot = px.bar(to_data_resetindex, x='neighbourhood', y='average_house_value', color='average_house_value', facet_row='year'
            , height=1000,width=800,
            labels={
                     "average_house_value": "Avg. House Price",
                     "neighbourhood": "Neighbourhood",
                     "average_house_value": "Avg. House Price"
                 },title="Average House Price in Toronto per Neighbourhood")
    return average_house_value_snapshot_plot
    # YOUR CODE HERE!

    
def top_most_expensive_neighbourhoods():
    """Top 10 most expensive neighbourhoods."""
    neighbourhoods_housevalue = to_data[['neighbourhood','average_house_value']]
    neighbourhoods_housevalue_avg = neighbourhoods_housevalue.groupby('neighbourhood').mean()
    top10_eighbourhoods_housevalue = neighbourhoods_housevalue_avg.nlargest(10, 'average_house_value')
    
    exp_neighb_plot = top10_eighbourhoods_housevalue.hvplot.bar(x='neighbourhood', y='average_house_value', xlabel='Neighbourhood'
                                , rot=90, ylabel='Avg. House Price', title='10 Most Expensive Neighbourhoods in T.O'
                                , height=500,width=800, yformatter=BasicTickFormatter(use_scientific=False)) 
    
    
    
    return exp_neighb_plot
    
    
    
    
    
    # YOUR CODE HERE!

def sunburts_cost_analysis():
    """Sunburst chart to conduct a costs analysis of most expensive neighbourhoods in Toronto per year."""
    
    # YOUR CODE HERE!
    to_data_resetindex = to_data.reset_index()
    sunburts_cost_analysis_plot = px.sunburst(to_data_resetindex, path=['year', 'neighbourhood'], values='average_house_value',
                 color='shelter_costs_owned', hover_data=['shelter_costs_owned'],
                 color_continuous_scale='RdBu')
    return sunburts_cost_analysis_plot
    
    
```
